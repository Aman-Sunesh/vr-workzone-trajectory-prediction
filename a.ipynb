```python
from pathlib import Path
import json
import numpy as np
import pandas as pd

DATA_DIR = Path("data")

def find_file(stem_or_name):
    """
    Finds a file inside DATA_DIR even if extension is missing in the name.
    Example: find_file("build_info") can find build_info.json.
    """
    direct = DATA_DIR / stem_or_name
    if direct.exists():
        return direct
    
    matches = list(DATA_DIR.glob(stem_or_name + ".*"))
    if matches:
        return matches[0]
    
    raise FileNotFoundError(f"Could not find {stem_or_name} in {DATA_DIR.resolve()}")

print("Data folder:", DATA_DIR.resolve())
print("Top-level files/folders:")
for p in DATA_DIR.iterdir():
    print("-", p.name)
```

```python
build_info_path = find_file("build_info")

print("File:", build_info_path)

with open(build_info_path, "r", encoding="utf-8") as f:
    build_info = json.load(f)

build_info
```

```python
windows_path = find_file("original_dataset_windows")

print("File:", windows_path)

if windows_path.suffix.lower() in [".xlsx", ".xls"]:
    original_dataset_windows = pd.read_excel(windows_path)
elif windows_path.suffix.lower() == ".csv":
    original_dataset_windows = pd.read_csv(windows_path)
else:
    raise ValueError(f"Unsupported file type: {windows_path.suffix}")

print(original_dataset_windows.shape)
original_dataset_windows.head()
```

```python
def preview_file(path, n=5):
    print("\n" + "=" * 80)
    print("File:", path)
    
    suffix = path.suffix.lower()
    
    try:
        if suffix == ".csv":
            df = pd.read_csv(path)
            print("Shape:", df.shape)
            display(df.head(n))
        
        elif suffix in [".xlsx", ".xls"]:
            df = pd.read_excel(path)
            print("Shape:", df.shape)
            display(df.head(n))
        
        elif suffix == ".json":
            with open(path, "r", encoding="utf-8") as f:
                obj = json.load(f)
            display(obj if isinstance(obj, dict) else obj[:n])
        
        elif suffix == ".npz":
            data = np.load(path, allow_pickle=True)
            print("Keys:", data.files)
            for key in data.files:
                arr = data[key]
                print(f"{key}: shape={arr.shape}, dtype={arr.dtype}")
                print(arr[:n] if arr.ndim > 0 else arr)
        
        else:
            print("Skipping unsupported type:", suffix)
    
    except Exception as e:
        print("Could not preview:", e)

supported_exts = {".csv", ".xlsx", ".xls", ".json", ".npz"}

for folder in [rawdata_dir, data_final_dir]:
    print("\n\n### Folder:", folder)
    files = [p for p in folder.rglob("*") if p.is_file() and p.suffix.lower() in supported_exts]
    print("Supported files found:", len(files))
    
    for p in files[:10]:
        preview_file(p)
```

```python
from pathlib import Path
import pickle
import numpy as np

pkl_files = list(Path("data/data_final").rglob("*.pkl"))
p = pkl_files[0]

with open(p, "rb") as f:
    sample = pickle.load(f)

print("FILE:", p)

print("\nMETA:")
print(sample["meta"])

print("\nEGO_REF:")
print(sample["ego_ref"])

print("\nAGENTS KEYS:")
for k, v in sample["agents"].items():
    print("\nKEY:", k)
    print("TYPE:", type(v))
    if hasattr(v, "shape"):
        print("SHAPE:", v.shape)
        print("DTYPE:", v.dtype)
        print("PREVIEW:", v[:2] if v.ndim > 0 else v)
    else:
        print("VALUE:", v)

print("\nMAP KEYS:")
for k, v in sample["map"].items():
    print("\nKEY:", k)
    print("TYPE:", type(v))
    if hasattr(v, "shape"):
        print("SHAPE:", v.shape)
        print("DTYPE:", v.dtype)
    elif isinstance(v, list):
        print("LEN:", len(v))
        print("PREVIEW:", str(v[:2])[:800])
    else:
        print("VALUE:", str(v)[:800])
```
